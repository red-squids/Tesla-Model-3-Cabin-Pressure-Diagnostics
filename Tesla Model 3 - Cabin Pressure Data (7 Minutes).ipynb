```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.rcParams['figure.figsize'] = [12, 5]
```

```python
out_data = pd.read_excel('/Users/dange/Pressure 2022-07-31 17-28-03 out.xlsx')
return_data = pd.read_excel('/Users/dange/Pressure 2022-07-31 17-38-32 back.xlsx')
```

```python
plt.style.use('fivethirtyeight')
#plt.style.use('seaborn-deep')
```

```python
# out_x = out_data['Time (s)']
out_x = out_data['Time (s)']
# out_y = out_data['Pressure (hPa)']
out_y = out_data['Pressure (hPa)']
# return_x = return_data['Time (s)']
return_x = return_data['Time (s)']
# return_y = return_data['Pressure (hPa)']
return_y = return_data['Pressure (hPa)']

plt.plot(out_x, out_y, '-', label='Trip Out', linewidth=2)
plt.plot(return_x, return_y, ':', label='Trip Home', linewidth=2)
plt.xlabel('Time (s)')
plt.ylabel('Pressure (hPa)')
plt.tight_layout()
plt.grid(True)


plt.title("Pressure reading with air recirculation toggled On/Off")
plt.legend()
plt.show()
```

## General parameters:
* The ambient outside air pressure was between 1015 - 1016 hPa.
* The recordings were taken in the evening during the same hour (same temperature), consecutively from the same device.

## Observations:
The data points collected for each leg of the trip appeared to be an inversed mirror in that peaks and valleys were experienced in roughly the same points during each trip. 

## Speculation:
There is a possibility that there is a correlation to be found in acceleration affecting air pressure, or possibly an effect due to terrain. 

## Way forward
In this phase, only approximately seven minutes of data was collected leaving a small sample for analysis. A larger sample size will be needed to observe patterns. Originally only barometric pressure was collected leaving out key environmental factors that could affect environmental pressure such as acceleration and g-forces. In following data collections, those factors will be included.

## Potential concerns
The absence of temperature data is a concern as temperature has an effect on environmental pressure. To remediate this concern, data will observed through external record and recorded as singular data points throughout the day post data collection but prior to analysis.
